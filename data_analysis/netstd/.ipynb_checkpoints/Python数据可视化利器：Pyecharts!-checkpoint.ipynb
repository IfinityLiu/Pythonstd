```python
from pyecharts.charts import Bar
from pyecharts.globals import ThemeType  # 导入风格包
from pyecharts import options as opts

import pandas as pd
import numpy as np
```

```python
v1 = [5, 20, 36, 10, 75, 90]
v2 = [10, 25, 8, 60, 20, 80]
x = ['衬衫', '羊毛衫', '雪纺衫', '裤子', '高跟鞋', '袜子']
```

```python
bar = (
    Bar(opts.InitOpts(theme=ThemeType.ESSOS))
    .add_xaxis(x)  # 添加一个x轴，在Bar()中选择风格样式
    .add_yaxis('商家A', v1)  # 添加一个y轴
    .add_yaxis('商家B', v2)  # 添加第二个y轴

    # 设置全局参数
    # 添加一个缩放的方式
    .set_global_opts(
        title_opts=opts.TitleOpts(title="商场数据显示"),  # 设置title

        xaxis_opts=opts.AxisOpts(
            splitline_opts=opts.SplitLineOpts(is_show=True)
        ),  # 设置x轴

            yaxis_opts=opts.AxisOpts(
                splitarea_opts=opts.SplitAreaOpts(
                    is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1))
        ),  # 设置y轴

        toolbox_opts=opts.ToolboxOpts(is_show=True),  # 设置工具箱

        datazoom_opts=[opts.DataZoomOpts(
            range_start=10, range_end=80, is_zoom_lock=False)],
    )

    # 设置系列参数
    # 设置辅助线
    .set_series_opts(
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_='max', name='最大值'),
                opts.MarkPointItem(type_='min', name='最小值'),
            ]
        ),

        markline_opts=opts.MarkLineOpts(
            data=[
                opts.MarkLineItem(type_='average', name='平均值')
            ]
        )
    )
)
bar.render_notebook()  # 在线显示
```

```python
#  折线图

from pyecharts.charts import Line

x1 = np.arange(100)
v11 = np.random.randn(100)*10
v12 = np.random.randn(100)*10

line = (
    Line()
    .add_xaxis(x)
    .add_yaxis('商家A', v1,is_smooth=True)  # is_smooth设置曲线是平滑的样式
    .add_yaxis('商家B', v2,is_step=True)  # is_step设置曲线是阶梯的样式
)

line.render_notebook()
```

```python
# 日历图

import datetime
import random
from pyecharts.charts import Calendar

begin = datetime.date(2017, 1, 1)
end = datetime.date(2017, 12, 31)
data = [
    [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
    for i in range((end - begin).days + 1)
]

c = (
    Calendar()
    .add("", data, calendar_opts=opts.CalendarOpts(range_="2017"))
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            max_=20000,
            min_=500,
            orient="horizontal",
            is_piecewise=True,
            pos_top="230px",
            pos_left="100px",
        ),
    )
)
c.render_notebook()
```

```python
# 散点图

from pyecharts.charts import Scatter

scal = (
    Scatter()
    .add_xaxis(x)
    .add_yaxis('商家A', v1)  # is_smooth设置曲线是平滑的样式
    .add_yaxis('商家B', v2)
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            type_='size', # 此处除了size还可以用color来设置
#             type_ = 'color',
            
            max_=150, min_=20
        )
    )
)

scal.render_notebook()
```

```python
from pyecharts.charts import EffectScatter
from pyecharts.globals import SymbolType  # 标记样式

sca2 = (
    EffectScatter(opts.InitOpts(width='1000px', height='500px'))
    .add_xaxis(x)
    .add_yaxis('商家A', v1, is_selected=True, symbol_size=10, symbol=SymbolType.ARROW)
    .add_yaxis('商家B', v2, is_selected=True, symbol_size=30)
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            type_='color',
            max_=150,
            min_=20,
        )
    )
)

sca2.render_notebook()
```
